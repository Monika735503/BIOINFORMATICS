```python
# ===============================
# STEP 0 — Install required tools
# ===============================
!apt-get update
!apt-get install -y default-jre fastqc velvet sra-toolkit

# ===============================
# STEP 1 — Download bacterial reads
# ===============================
!mkdir -p pg_reads
%cd pg_reads

# Download paired-end reads from SRA (Porphyromonas gingivalis)
!fastq-dump --split-files SRR5253683

# Go back to main directory
%cd ..

# ===============================
# STEP 2 — Run FastQC on reads
# ===============================
import os
from google.colab import files

# Run FastQC on read 1
!fastqc pg_reads/SRR5253683_1.fastq
files.download('pg_reads/SRR5253683_1_fastqc.html')

# Run FastQC on read 2
!fastqc pg_reads/SRR5253683_2.fastq
files.download('pg_reads/SRR5253683_2_fastqc.html')

# ===============================
# STEP 3 — Prepare assembly directories
# ===============================
!mkdir -p assembly_module/data

# Move FASTQ files into assembly folder
!mv pg_reads/SRR5253683_1.fastq assembly_module/data/
!mv pg_reads/SRR5253683_2.fastq assembly_module/data/

# ===============================
# STEP 4 — Run Velvet Assembly
# ===============================

# Step 4a: velveth — initialize hash table
!velveth PG_ASSEMBLY 31 -shortPaired \
    -fastq assembly_module/data/SRR5253683_1.fastq \
    assembly_module/data/SRR5253683_2.fastq

# Step 4b: velvetg — build the assembly
!velvetg PG_ASSEMBLY -exp_cov auto -cov_cutoff auto

# ===============================
# STEP 5 — View and download results
# ===============================
print("Files in PG_ASSEMBLY folder:")
print(os.listdir('PG_ASSEMBLY'))

# Download contigs
files.download('PG_ASSEMBLY/contigs.fa')

# Download assembly statistics
files.download('PG_ASSEMBLY/stats.txt')
```
