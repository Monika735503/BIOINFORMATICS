```python
import os
from google.colab import files
!pip install biopython
from Bio import SeqIO, Entrez
Entrez.email = "monikav.bt23@rvce.edu.in"

mysearch = Entrez.esearch(db='nucleotide', term="[Alzheimer disease]")
myrecord = Entrez.read(mysearch)
mysearch.close()
myids = myrecord["IdList"]

for id1 in myids:
  filename = id1 + ".fasta"
  if not os.path.isfile(filename):
    net_handle = Entrez.efetch(
        db='nucleotide', id=id1, rettype="fasta", retmode="text"
    )
    out_handle = open(filename, "w")
    out_handle.write(net_handle.read())
    files.download(filename)
    out_handle.close()
    net_handle.close()

  myrecord = SeqIO.read(filename, "fasta")
  print(myrecord)

  # Convert DNA → RNA
  rna_seq = myrecord.seq.transcribe() #Uses Biopython’s built-in method to convert DNA to RNA automatically (T → U)
  print("\nRNA Sequence:\n", rna_seq)
```
